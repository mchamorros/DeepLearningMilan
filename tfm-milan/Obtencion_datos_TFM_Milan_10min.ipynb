```python
#Referencia a todas las  librerias usadas
import pandas as pd
import datetime
import sys
import numpy
import json
import numpy as np
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon
import pandas as pd
import datetime
```

```python
parse = lambda x: datetime.datetime.fromtimestamp(float(x)/1000)
matches = ['5459', '6372', '5644', '6377', '5570', '6561']

sliceDataCells = pd.DataFrame({})
```

```python
#Obtención de datos de los CellId indicados anteriormente de Noviembre

for index in range(1,31):
    sliceSum2 = pd.read_csv('original_data/milan_tel/sms-call-internet-mi-2013-11-'+str(index).zfill(2) +'.txt', sep='\t', encoding="utf-8-sig", names=['CellID', 'datetime', 'countrycode', 'smsin', 'smsout', 'callin', 'callout', 'internet'], parse_dates=['datetime'], date_parser=parse)
    sliceSum2 = sliceSum2.drop(sliceSum2[~(sliceSum2['CellID'].isin(matches))].index)
    sliceSum2 = sliceSum2.set_index('datetime')
    sliceSum2['minute'] = sliceSum2.index.minute
    sliceSum2['hour'] = sliceSum2.index.hour
    sliceSum2['weekday'] = sliceSum2.index.weekday
    sliceSum2['week'] = sliceSum2.index.isocalendar().week
    
    sliceSum2 = sliceSum2.groupby(['week', 'weekday', 'hour', 'minute', 'CellID'], as_index=False).sum() #Elimino countrycode de los campos.
    
    
    #(any(sliceSum2['CellID'] in s for s in matches))
    #| sliceSum2['CellID'] != 6372 | sliceSum2['CellID'] != 5644 | sliceSum2['CellID'] != 5570 | sliceSum2['CellID'] != 6561
    sliceDataCells = sliceDataCells.append(sliceSum2)

sliceDataCells['idx'] = (sliceDataCells['minute']/10) + sliceDataCells['hour']*6 + (sliceDataCells['weekday']*24*6) + (sliceDataCells['week']*24*7*6)
sliceDataCells.head()
```

```python
#Obtención de datos de los CellId indicados anteriormente de Diciembre

for index in range(1,32):
    sliceSum2 = pd.read_csv('original_data/milan_tel/sms-call-internet-mi-2013-12-'+str(index).zfill(2) +'.txt', sep='\t', encoding="utf-8-sig", names=['CellID', 'datetime', 'countrycode', 'smsin', 'smsout', 'callin', 'callout', 'internet'], parse_dates=['datetime'], date_parser=parse)
    sliceSum2 = sliceSum2.drop(sliceSum2[~(sliceSum2['CellID'].isin(matches))].index)
    sliceSum2 = sliceSum2.set_index('datetime')
    sliceSum2['minute'] = sliceSum2.index.minute
    sliceSum2['hour'] = sliceSum2.index.hour
    sliceSum2['weekday'] = sliceSum2.index.weekday
    sliceSum2['week'] = sliceSum2.index.isocalendar().week
    
    sliceSum2 = sliceSum2.groupby(['week', 'weekday', 'hour', 'minute', 'CellID'], as_index=False).sum() #Elimino countrycode de los campos.
    
    #(any(sliceSum2['CellID'] in s for s in matches))
    #| sliceSum2['CellID'] != 6372 | sliceSum2['CellID'] != 5644 | sliceSum2['CellID'] != 5570 | sliceSum2['CellID'] != 6561
    sliceDataCells = sliceDataCells.append(sliceSum2)

sliceDataCells['idx'] = (sliceDataCells['minute']/10) + sliceDataCells['hour']*6 + (sliceDataCells['weekday']*24*6) + (sliceDataCells['week']*24*7*6)
sliceDataCells.head()
```

```python
matches = ['Relative Humidity', 'Precipitation', 'Temperature', 'Wind Speed']
pdDataMeteoCellId = pdDataMeteoCellId.drop(pdDataMeteoCellId[~(pdDataMeteoCellId['sensorType'].isin(matches))].index)
```

```python
pdDataMeteoCellId.loc[pdDataMeteoCellId['sensorType'] == 'Relative Humidity', 'sensorType'] = 'relativeHumidity'
pdDataMeteoCellId.loc[pdDataMeteoCellId['sensorType'] == 'Precipitation', 'sensorType'] = 'precipitation'
pdDataMeteoCellId.loc[pdDataMeteoCellId['sensorType'] == 'Temperature', 'sensorType'] = 'temperature'
pdDataMeteoCellId.loc[pdDataMeteoCellId['sensorType'] == 'Wind Speed', 'sensorType'] = 'windSpeed'
```
